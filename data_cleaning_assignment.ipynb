1. Run df.isnull().sum() to count missing values per column.

**1. count missing values per column**

```python
df.isnull().sum()
```

**2.Numeric columns** :(age, salary, price, weight_kg): fill with median.

```python
dfFirstCopy = df.copy()
```

```python
dfFirstCopy.info()
```

There are some columns that are object,so firstly we need to change it

```python
col_to_change = ['age', 'salary', 'price', 'weight_kg']
for col in col_to_change:
  dfFirstCopy[col] = pd.to_numeric(dfFirstCopy[col], errors='coerce')
```

now convert all null value to median

```python
col_to_fill = ['age', 'salary', 'price', 'weight_kg']
for col in col_to_fill:
  dfFirstCopy[col].fillna(dfFirstCopy[col].median(), inplace=True)
```

check ->is it solve the numeric columns null value problem

```python
dfFirstCopy.isnull().sum()
```

**3.Categorical columns**: (department, gender, country, city): fill with mode.

```python
categorical_col = ['department','gender','country','city']
for col in categorical_col:
  dfFirstCopy[col] = dfFirstCopy[col].fillna(dfFirstCopy[col].mode()[0])
```

check ->is it solve the categorical columns null value problem

```python
dfFirstCopy.isnull().sum()
```

**4. Date column (join_date): forward fill or fill with most frequent date.**

```python
dfFirstCopy.info()
```

join_date column data type was object thats why i need to change it

```python
dfFirstCopy['join_date'] = pd.to_datetime(dfFirstCopy['join_date'], format='mixed', dayfirst=True)
```

fill it with forward fill -> ffill

```python
dfFirstCopy['join_date'] = dfFirstCopy['join_date'].ffill()
```

check ->is it solve the date columns null value problem

```python
dfFirstCopy.isnull().sum()
```

**5. review column** : fill with 'No Review'

```python
dfFirstCopy['review'] = dfFirstCopy['review'].fillna('No Review')
```

check ->is it solve the review column null value problem

```python
dfFirstCopy.isnull().sum()
```

**6. is_active:** fill with False (conservative approach).

```python
dfFirstCopy['is_active'] = dfFirstCopy['is_active'].fillna(False)
```
